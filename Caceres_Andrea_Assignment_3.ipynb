## Q1 (10 pt) Factorial number




Create two functions to calculate a factorial. One recursive and one iterative. To ensure that your functions are functioning properly, please follow the test procedure outlined at the end of this section.

**Definition of a Factorial**
The factorial of a number is the multiplication of all the numbers between 1 and the number itself. It is written like this: n!. So the factorial of 2 is 2! (= 1 × 2).


To calculate a factorial you need to know two things:

```bash
0! = 1
1! = 1
n! = (n - 1)! × n
```
The factorial of 0 has value of 1, and the factorial of a number n is equal to the multiplication between the number n and the factorial of n-1.

Q1.1 What is the diffference between a recursive and iterative process?

Q1.1 Response Here

````python
def factorial_recursion(n):
  #handling strings
  if isinstance(n, str):
    return math.nan
    #handling numbers less than 0
  if n<0:
    return math.nan
    #when n is 0 and 1
  elif ((n==0) | (n==1)):
    return 1
  else:
    return n*factorial_recursion(n-1)
````

````python
def factorial_iterative(n):
  #handling strings
  if isinstance(n, str):
    return math.nan
    #handling numbers less than 0
  if n<0:
    return math.nan
  elif ((n==0) | (n==1)):
    return 1
  result = 1
  #for loop for numbers 1:n+1, taking the result of the factorial and mulitply by i
  for i in range (1, n+1):
    result *= i
  return result
````

A recursive proccess calls on itself while an iterative proccess is made of loops.

## Q5 (20 pt.) Find common elements


Q5.1 Write a function to find common element in three sorted arrays

Q5.2 Include a flow chart of your solution

Q5.3 Discuss the complexity of your solution (Big O Notation)

Note: Given three arrays sorted in non-decreasing order, print all common elements in these arrays

**Sample Input**

input1 = [1, 5, 10, 20, 40, 80]

input2 = [6, 7, 20, 80, 100]

input3 = [3, 4, 15, 20, 30, 70, 80, 120]


**Sample Output**

Output: [20, 80]

````python
#Q5.1
def commonElements(array1,array2,array3):
  #setting the counters
  x=0
  x2=0
  x3=0
  common=[] #making an empty list for the common numbers

  while x<len(array1) and x2<len(array2) and x3<len(array3):
    if array1[x] == array2[x2] == array3[x3]:
      common.append(array1[x]) #append when the values are equal at element
      x+=1
      x2+=1
      x3+=1
    elif array1[x]<array2[x2]:
      #adding one to check the next element
      x+=1
    elif array2[x2]<array3[x3]:
      x2+=1
    else:
      x3+=1
#return the common list with appended common values
  return(common)
array1 = [1,5,10,20,40,80]
array2=[6,7,20,80,100]
array3 = [3,4,15,20,30,70,80,120]

commonElements(array1,array2,array3)
````

Q6.1 You are given an unordered list of unique integers `nums` in the range [1,n], where `n` represent the length of `nums + 2`. This means that two numbers in this range are missing from the list.

write a function that takes in this list and return a new list with the two missing numbers, sorted numerically.

Q6.2 Include a flow chart of your solution.

Q6.3 Discuss the complexity of your solution (Big O Notation)

**Sample Input**

nums = [1,4,3]

Sample Output

[2,5]


````python
def missingNumbers(nums):
  # n from the problem
  n = len(nums) + 2
  current = [False] * (n+1)
  for num in nums:
    current[num] = True
    #make an empty list for missing values
    missing = []
    #loop to go through each value and seeing if that valus is in the list, if not add to missing list
    for x in range(1,n+1):
      if current[x] == False:
        missing.append(x)
  return missing
nums = [1,4,3]
missingNumbers(nums)
  # list of numbers 1:n
  # return the missing values
````
