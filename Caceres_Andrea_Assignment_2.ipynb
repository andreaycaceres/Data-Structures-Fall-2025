##Description of My Approach

For my first algorithm, which will be my slowest, I will be conducting a linear search of the random number. I will select a random number between one of the three ranges (1:100), (1:10,000) and (1: 10,000,000). Then the function will guess numbers one-by-one until the correct number is found. For my second algorithm, I will be using the binary approach. The binary search will cut the range of potential numbers in half by asking the user if the number is less than or greater than the middle number of the range. Then, the function will proceed to split the range in half until the range is small enough and then choose one-by-one till the number is found.

##Code: Linear Approach

```python
#Algorithm 1
import random
import time

def linear_approach():
  low = int(input("Enter the lowest number in your range:"))
  high = int(input("Enter the highest number in your range:"))

  print("Think of a number between the range selected:")

  attempts = 0
  start_time = time.time()

  for guess in range(low, high +1):
    print(f"Is your number {guess}? (Y/N):")
    user = input().strip().lower()
    attempts +=1

    if user == 'y':
      print(f"Yay I got it right in {attempts} attempts!")
      end_time=time.time()
      totaltime = round(end_time-start_time, 2)
      print(f"Total time is {totaltime} seconds")
      break


linear_approach()






```

##BINARY SEARCH: CODE

```python
import time
def binary_search():
  low = int(input("Enter the lowest number in your range:"))
  high = int(input("Enter the highest number in your range:"))

  print("Think of a number between the range:")

  attempts = 0
  begintime = time.time()

  while low<=high:
      middle = (low+high)//2
      print(f"Is the answer {middle}? (Y/N):")
      user = input().strip().lower()

      attempts += 1

      if user == 'y':
        print(f"Yay I got it right in {attempts} attempts!")
        break

      elif user == 'n':
          print(f"is the number greater than {middle} (Y/N):")
          higher = input().strip().lower()

      if higher == "y":
              low = middle +1
      elif higher == 'n':
              high = middle -1

  endtime =time.time()
  totaltime = round(endtime-begintime, 2)
  print(f"Total time is {totaltime} seconds")

binary_search()

```

```python
#Code to collect average for linear search
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
import random

def linear_simulation(low, high, trials):
  results=[]

  for _ in range(trials):
    answer = random.randint(low, high)
    attempts = 0
    start_time = time.time()

    for guess in range(low, high+1):
      attempts += 1
      if guess == answer:
          break
      endtime = time.time()
      totaltime=endtime-start_time

      results.append({'attempts':attempts, 'correct answer':answer, 'total time':totaltime})

    df = pd.DataFrame(results)
    return df
```

```python
#Code to collect averages for binary search
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
import random

def binary_data(low, high, answer):
  attempts = 0
  start_time= time.time()

  while low<= high:
    middle = (low+high)//2
    attempts +=1

    if middle == answer:
      break
    elif middle < answer:
      low = middle +1
    else:
      high = middle -1
  endtime = time.time()
  totaltime = endtime - start_time
  return attempts, totaltime

def binary_simulation(low, high, trials=100):
  results = []
  for _ in range(trials):
    answer = random.randint(low, high)
    attempts, totaltime = binary_data(low, high, answer)

    results.append({'correct answer':answer, 'attempts taken': attempts, 'total time':totaltime})
  df = pd.DataFrame(results)
  return df
```

##Discussion & Conclusion:

The binary search will always be faster than the linear search because with each attempt the range of potential correct answers is split in half. Doing this will result in finding the answers much faster rather than choosing numbers one by one. The size of the range impacts performance because the larger the number gets, the more attempts have to be made to split the range down enough to make a good educated guess. The odds of finding the answer in less attempts and faster time are higher when the range of possibilities is smaller. A way algorithm efficiency shows up in the real-world is in the health industry.
